```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('C:\\Users\\Mamdouh\\Downloads\\drug200.xls')
df
```

```python
df.isnull().sum()
```

```python
dummies=pd.get_dummies(df.Sex)
dummies_1=pd.get_dummies(df.BP)
dummies_2=pd.get_dummies(df.Cholesterol)
```

```python
df_dummies=pd.concat([df,dummies,dummies_1,dummies_2] , axis='columns')
df_dummies
```

```python
from sklearn.preprocessing import LabelEncoder
df_dummies['Drug']=LabelEncoder().fit_transform(df_dummies['Drug'])
df_dummies['Drug']
```

```python
df_dummies
```

```python
df=df_dummies.drop(['Sex', 'BP', 'Cholesterol'] ,axis=1)
df
```

```python
df.info
```

```python
x=df.drop(['Drug'],axis=1)
x
```

```python
y=df.Drug
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y ,test_size=0.2)
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=5)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```
